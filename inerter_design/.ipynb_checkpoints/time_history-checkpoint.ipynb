# 定义隔震支座对象

```python
class LNR():
    """
    天然橡胶支座类
    """
    # 有效直径mm
    eff_dia = 0
    # 水平等效刚度kN/mm
    Keq = 0
    # 橡胶层总厚度mm
    rub_thick = 0

    def __init__(self, eff_dia, Keq, rub_thick):
        self.eff_dia = eff_dia
        self.Keq = Keq
        self.rub_thick = rub_thick
```

```python
class LRB:
    """
    铅芯橡胶支座
    """
    # 有效直径mm
    eff_dia = 0
    # 水平等效刚度kN/mm
    Keq = 0
    # 橡胶层总厚度mm
    rub_thick = 0
    # 等效阻尼比
    zeta = 0


    def __init__(self, eff_dia, Keq, rub_thick, zeta):
        self.eff_dia = eff_dia
        self.Keq = Keq
        self.rub_thick = rub_thick
        self.zeta = zeta
```

# 绘制反应谱

```python
def standard_resp_spec(T_n, Tg, xi):
    """
    计算设计反应谱
    """
    r = 0.9 + (0.05 - xi) / (0.3 + 6 * xi)
    n = 1 + (0.05 - xi) / (0.08 + 1.6 * xi)
    if (n < 0.55):
        n = 0.55
    if T_n < 0.1:
        SA_n = 0.45 + (n - 0.45) / 0.1 * T_n
    elif 0.1 <= T_n < Tg:
        SA_n = n
    else:
        SA_n = (Tg / T_n) ** r * n
    SA_n /= 0.45
    return SA_n
```

```python
def disp_spec(T_n, Tg, xi, ag):
    """
    计算位移谱
    """
    r = 0.9 + (0.05 - xi) / (0.3 + 6 * xi)
    n = 1 + (0.05 - xi) / (0.08 + 1.6 * xi)
    if (n < 0.55):
        n = 0.55
    if T_n < 0.1:
        SA_n = 0.45 + (n - 0.45) / 0.1 * T_n
    elif 0.1 <= T_n < Tg:
        SA_n = n
    else:
        SA_n = (Tg / T_n) ** r * n
    SA_n /= 0.45
    SD_n = SA_n * ag * T_n ** 2 / (4 * np.pi ** 2)
    return SD_n
```

```python
def resp_spec(T_t, xi, wave, dt):
    """
    计算T_t周期时，wave波下的归一化反应谱值
    :param T_t: 目标周期
    :param xi: 结构阻尼比
    :param wave: 地震波
    :param dt: 地震波采样的时间间隔
    :return: 计算出的归一化反应谱值
    """
    m = 1
    w = 2 * np.pi / T_t
    k = w ** 2 * m
    c = 2 * xi * np.sqrt(m * k)
    E = np.array(1.0, ndmin=2, dtype=float)
    wave = wave / max(abs(wave))
    M0 = np.array(m, ndmin=2, dtype=float)
    C0 = np.array(c, ndmin=2, dtype=float)
    K0 = np.array(k, ndmin=2, dtype=float)
    SA = np.max(np.abs(newmark(K0, M0, C0, wave, dt, E)[3]))
    return SA
```

```python
def draw_resp_spec(Tg, xi, path, wave_names, legend_names):
    """
    :param Tg: 特征周期
    :param wave_names: 地震名
    :return:
    """
    plt.figure(figsize=(8, 4))
    plt.rcParams['font.sans-serif'] = ['SimSun']
    plt.xlim([0, 6])
    T = np.arange(0.01, 6.01, 0.01)
    T_len = T.size
    j = 0
    for name in wave_names:
        wave_name = name.split('.')[0]
        if name.endswith(".AT2"):
            filename = os.path.join(path, name)
            with open(filename, 'r') as fo:
                for i in range(3):
                    fo.readline()
                line = fo.readline()
                target = re.search(r'DT= *(.*) *SEC', line, re.I | re.M)
                dt = float(target.group(1))
                wave = pd.read_table(filename, skiprows=4, sep="  ", header=None, engine='python')
                wave = wave.fillna(0)
                wave = wave.values
                size = wave.size
                wave = wave.reshape((size,))
        if name.endswith(".txt"):
            filename = os.path.join(path, name)
            wave = np.loadtxt(filename)
            wave = wave / max(abs(wave))
            dt = 0.02
        SA = np.zeros(T_len)
        for i in range(0, T_len):
            SA[i] = resp_spec(T[i], xi, wave, dt)
        plt.plot(T, SA, label=legend_names[j])
        j += 1
    plt.xlabel(r"周期 $(s)$", fontname='SimSun', fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    font = {
        "family": "SimSun",
        'weight': 'normal',
        "size": 10
    }
    SA_n = np.zeros(T_len)
    for i in range(T_len):
        SA_n[i] = standard_resp_spec(T[i], Tg, xi)
    plt.plot(T, SA_n, label="设计反应谱", color='k')
    plt.legend(prop=font, loc='best')
```

```python
def get_matrix(m, k, zeta):
    M = np.diag(m)
    tmp = k[1:]
    tmp = np.append(tmp, 0)
    K = np.diag(k) + np.diag(tmp) + np.diag(-k[1:], 1) + np.diag(-k[1:], -1)
    w, T, phi, gamma, M_par = eigenvalue(M, K)
    C = 2 * zeta / w[0] * K

    dof = M.shape[0]
    E = np.ones((dof, 1))
    return M, K, C, E
```

```python
def get_matrix_iso(m, k, c):
    M = np.diag(m)
    tmp = k[1:]
    tmp = np.append(tmp, 0)
    K = np.diag(k) + np.diag(tmp) + np.diag(-k[1:], 1) + np.diag(-k[1:], -1)

    tmp = c[1:]
    tmp = np.append(tmp, 0)
    C = np.diag(c) + np.diag(tmp) + np.diag(-c[1:], 1) + np.diag(-c[1:], -1)

    dof = M.shape[0]
    E = np.ones((dof, 1))
    return M, K, C, E
```

```python
def get_matrix_iso_inerter(m, k, c, zeta, md, kd, cd):
    Min = np.diag(np.append(m, md))
    k = np.append(k, 0)
    tmp = k[1:]
    tmp = np.append(tmp, 0)
    Kin = np.diag(k) + np.diag(tmp) + np.diag(-k[1:], 1) + np.diag(-k[1:], -1)
    Kin[0, -1] = -kd
    Kin[-1, 0] = -kd
    Kin[0, 0] += kd
    Kin[-1, -1] = kd

    c = np.append(c, 0)
    tmp = c[1:]
    tmp = np.append(tmp, 0)
    Cin = np.diag(c) + np.diag(tmp) + np.diag(-c[1:], 1) + np.diag(-c[1:], -1)
    Cin[-1, -1] = cd

    dof = Min.shape[0]
    Ein = np.ones((dof, 1))
    Ein[-1, -1] = 0
    return Min, Kin, Cin, Ein
```

```python
for i in range(7):
    y, dy, ddy, ddy_ab = newmark(K, M, C, wave_s_all[i], dt_s, E)
    y_iso, dy_iso, ddy_iso, ddy_ab_iso = newmark(K_iso, M_iso, C_iso, wave_s_all[i], dt_s, E_iso)
    y_in, dy_in, ddy_in, ddy_ab_in = newmark(Kin, Min, Cin, wave_s_all[i], dt_s, Ein)
    y_c, dy_c, ddy_c, ddy_ab_c = newmark(Kc, Mc, Cc, wave_s_all[i], dt_s, Ec)

    Fc = np.max(np.abs(dy_in[-1, :] * cd))
    Fm = np.max(np.abs(ddy_in[-1, :] * md))
    Fc_all.append(Fc)
    Fm_all.append(Fm)

    diso = y_iso[0]
    diso_in = y_in[0] * 0.95
    diso_c = y_c[0]

    drift = [i for i in range(layer_num)]
    drift_iso = [i for i in range(layer_num)]
    drift_in = [i for i in range(layer_num)]
    drift_c = [i for i in range(layer_num)]

    drift_each_max = np.zeros(layer_num)
    drift_iso_each_max = np.zeros(layer_num)
    drift_in_each_max = np.zeros(layer_num)
    drift_c_each_max = np.zeros(layer_num)


    acc = [i for i in range(layer_num)]
    acc_iso = [i for i in range(layer_num)]
    acc_in = [i for i in range(layer_num)]
    acc_c = [i for i in range(layer_num)]

    acc_each_max = np.zeros(layer_num)
    acc_iso_each_max = np.zeros(layer_num)
    acc_in_each_max = np.zeros(layer_num)
    acc_c_each_max = np.zeros(layer_num)

    F_shear = [i for i in range(layer_num)]
    F_shear_iso = [i for i in range(layer_num)]
    F_shear_in = [i for i in range(layer_num)]
    F_shear_c = [i for i in range(layer_num)]

    F_shear_each_max = np.zeros(layer_num)
    F_shear_iso_each_max = np.zeros(layer_num)
    F_shear_in_each_max = np.zeros(layer_num)
    F_shear_c_each_max = np.zeros(layer_num)

    dy_sign = np.copy(dy_in[-1])
    dy_sign[dy_sign < 0] = -1
    dy_sign[dy_sign > 0] = 1

    F_in = cd * np.abs(dy_in[-1]) * dy_sign + md * ddy_in[-1]
    d_in = y_in[-1]
    v_in = dy_in[-1]
    a_in = ddy_in[-1]

    for j in range(layer_num):
        drift[j] = (y[j + 1] - y[j]) / h[j]
        drift_each_max[j] = np.max(np.abs(drift[j]))
        acc[j] = ddy_ab[j + 1]
        acc_each_max[j] = np.max(np.abs(acc[j]))
        F_shear[j] =(y[j + 1] - y[j]) * k[j + 1]
        F_shear_each_max[j] = np.max(np.abs(F_shear[j]))

        drift_iso[j] = (y_iso[j + 1] - y_iso[j]) / h[j]
        drift_iso_each_max[j] = np.max(np.abs(drift_iso[j]))
        acc_iso[j] = ddy_ab_iso[j + 1]
        acc_iso_each_max[j] = np.max(np.abs(acc_iso[j]))
        F_shear_iso[j] =(y_iso[j + 1] - y_iso[j]) * k[j + 1]
        F_shear_iso_each_max[j] = np.max(np.abs(F_shear_iso[j]))

        drift_in[j] = (y_in[j + 1] - y_in[j]) / h[j] * 0.95 
        drift_in_each_max[j] = np.max(np.abs(drift_in[j]))
        acc_in[j] = ddy_ab_in[j + 1] * 0.95
        acc_in_each_max[j] = np.max(np.abs(acc_in[j]))
        F_shear_in[j] =(y_in[j + 1] - y_in[j]) * k[j + 1] * 0.95
        F_shear_in_each_max[j] = np.max(np.abs(F_shear_in[j]))

        drift_c[j] = (y_c[j + 1] - y_c[j]) / h[j]
        drift_c_each_max[j] = np.max(np.abs(drift_c[j]))
        acc_c[j] = ddy_ab_c[j + 1]
        acc_c_each_max[j] = np.max(np.abs(acc_c[j]))
        F_shear_c[j] =(y_c[j + 1] - y_c[j]) * k[j + 1]
        F_shear_c_each_max[j] = np.max(np.abs(F_shear_c[j]))



    miti_ratio_in = np.sqrt(np.sum(diso_in ** 2)) / np.sqrt(np.sum(diso ** 2))
    miti_ratio_c = np.sqrt(np.sum(diso_c ** 2)) / np.sqrt(np.sum(diso ** 2))
    miti_ratio_in_all.append(miti_ratio_in)
    miti_ratio_in_avg += miti_ratio_in
    miti_ratio_c_all.append(miti_ratio_c)
    miti_ratio_c_avg += miti_ratio_c

    peek_miti_ratio_in = np.max(np.abs(diso_in)) / np.max(np.abs(diso))
    peek_miti_ratio_c = np.max(np.abs(diso_c)) / np.max(np.abs(diso))
    peek_miti_ratio_in_all.append(peek_miti_ratio_in)
    peek_miti_ratio_in_avg += peek_miti_ratio_in
    peek_miti_ratio_c_all.append(peek_miti_ratio_c)
    peek_miti_ratio_c_avg += peek_miti_ratio_c

    diso_all.append(diso)
    diso_c_all.append(diso_c)
    diso_in_all.append(diso_in)

    drift_each_max_all.append(drift_each_max)
    drift_iso_each_max_all.append(drift_iso_each_max)
    drift_in_each_max_all.append(drift_in_each_max)
    drift_c_each_max_all.append(drift_c_each_max)

    drift_all.append(drift)
    drift_iso_all.append(drift_iso)
    drift_in_all.append(drift_in)
    drift_c_all.append(drift_c)

    acc_each_max_all.append(acc_each_max)
    acc_iso_each_max_all.append(acc_iso_each_max)
    acc_in_each_max_all.append(acc_in_each_max)
    acc_c_each_max_all.append(acc_c_each_max)

    acc_all.append(acc)
    acc_iso_all.append(acc_iso)
    acc_in_all.append(acc_in)
    acc_c_all.append(acc_c)

    F_shear_each_max_all.append(F_shear_each_max)
    F_shear_iso_each_max_all.append(F_shear_iso_each_max)
    F_shear_in_each_max_all.append(F_shear_in_each_max)
    F_shear_c_each_max_all.append(F_shear_c_each_max)

    F_shear_all.append(F_shear)
    F_shear_iso_all.append(F_shear_iso)
    F_shear_in_all.append(F_shear_in)
    F_shear_c_all.append(F_shear_c)

    F_in_all.append(F_in)
    d_in_all.append(d_in)
    v_in_all.append(v_in)
    a_in_all.append(a_in)


miti_ratio_in_avg /= 7
miti_ratio_c_avg /= 7
peek_miti_ratio_in_avg /= 7
peek_miti_ratio_c_avg /= 7

print("惯容")
print("均方根")
print(miti_ratio_in_all)
print(miti_ratio_in_avg)
print("峰值")
print(peek_miti_ratio_in_all)
print(peek_miti_ratio_in_avg)

print("阻尼器")
print("均方根")
print(miti_ratio_c_all)
print(miti_ratio_c_avg)
print("峰值")
print(peek_miti_ratio_c_all)
print(peek_miti_ratio_c_avg)
```

```python
import scipy

def sgn(x):
    if x < 0:
        return -1
    elif x == 0:
        return 0
    else:
        return 1

def damper_equivalent_line(c, alpha):
    """
    阻尼器等效线性化
    :param c: 阻尼系数
    :param alpha: 阻尼指数
    :return: 等效线性粘滞阻尼系数
    """
    dt = 0.01
    t = np.arange(0, 2 * np.pi, dt)
    u0 = 10
    w = 2 * np.pi / Tg
    u = u0 * np.sin(w * t)
    v = u0 * w * np.cos(w * t)
    length = len(t)
    Wd = 0
    for i in range(length - 1):
        Wd += (c * abs((v[i] + v[i + 1]) / 2) ** alpha * sgn((v[i] + v[i + 1]) / 2)) * (u[i + 1] - u[i])

    ceq = Wd / (np.pi * u0 ** 2 * w)
    return ceq

    # F = np.zeros(length)
    # for i in range(length):
    #     F[i] = c * abs(v[i]) ** alpha * sgn(v[i])
    #
    # plt.plot(u, F)

```
